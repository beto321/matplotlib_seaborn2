```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
country=['India','Japan','Mexico','Colombia','Germany']
population=[1000, 800, 900, 1000, 300]
```

```python
#barplot: Nice for categorical variables
plt.bar(country,population,width=0.5,color=['red','blue'])
plt.xticks(np.arange(5),('Indi','Japon','Mexicoxlan','Colombian','Alemania'),rotation=45)
plt.show()
```

```python
#barplot in a horizontal way
plt.barh(country,population)
plt.show()
```
