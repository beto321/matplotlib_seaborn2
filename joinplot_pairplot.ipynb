```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips =sns.load_dataset('tips')
tips.head(2)
```

```python
#we can see the relation between two numerical variables using two different plots
sns.jointplot(data=tips, x='total_bill',y='tip')
plt.show()
```

```python
#we can customize it even more.  Now we will add a third variable: sex
sns.jointplot(data=tips, x='total_bill',y='tip', hue='sex')
plt.show()
```

```python
#we can choose the type of plot using kind and marginal_ticks, y marginal kws
sns.jointplot(data=tips, x='total_bill',y='tip', hue='sex', kind='hist', marginal_ticks=True, marginal_kws=dict(bins=25, fill=False, multiple='dodge'))
plt.show()
```

## Vemos Pairplot

```python
# pairplot works for numerical variables
# the main diagonal will always be an histogram
sns.pairplot(data=tips)
plt.show()
```

```python
#we can add a third variable, like sex (categorical variable)
sns.pairplot(data=tips, hue='sex')
plt.show()
```

```python
# we can work with the palette
sns.pairplot(data=tips, hue='sex',palette='coolwarm')
plt.show()
```

```python
# with corner, we avoid duplicate information
sns.pairplot(data=tips, hue='sex',palette='coolwarm', corner=True)
plt.show()
```

```python
# we can also use kind
sns.pairplot(data=tips, hue='sex',palette='coolwarm', corner=True, kind='scatter')
plt.show()
```
