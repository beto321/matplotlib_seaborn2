```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips=sns.load_dataset('tips')
tips.head(2)
```

```python
sns.countplot(data=tips,x='day',hue='sex')
plt.show()
```

```python
sns.stripplot(data=tips, x='day',y='total_bill',hue='sex', dodge=True)
plt.show()
```

```python
#we can configure the figure
plt.figure(figsize=(6,6))
sns.stripplot(data=tips, x='day',y='total_bill',hue='sex', dodge=True)
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.swarmplot(data=tips, x='day',y='total_bill',hue='sex', dodge=True)
plt.show()
```

```python
#we are working with a categorical variable: day
plt.figure(figsize=(6,6))
sns.boxplot(data=tips, x='day',y='total_bill',hue='sex', dodge=True)
plt.show()
```

```python
sns.boxplot(data=tips,y='total_bill') # it works with numerical variables too
plt.show()
```

```python
#we can mix plots
plt.figure(figsize=(6,6))
sns.boxplot(data=tips, x='day',y='total_bill',hue='sex', dodge=True)
sns.swarmplot(data=tips, x='day',y='total_bill',hue='sex', dodge=True, color='0', marker='<')
plt.show()
```

```python
#configuration is very similar
plt.figure(figsize=(6,6))
sns.violinplot(data=tips, x='day',y='total_bill',hue='sex', split=True, palette='pastel')
plt.show()
```

```python
#violinplot also works with numerical variables
sns.violinplot(data=tips, y='total_bill')
plt.show()
```

```python
#the best one
#very similar to displot - It means you can choose wich plot you want in parameters
sns.catplot(data=tips, x='day',y='total_bill', hue='sex',dodge=True, kind='swarm',col='time')
plt.show()
```
