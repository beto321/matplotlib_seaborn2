```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.barplot(x=['A','B','C'],y=[1,3,2])
plt.show()
```

```python
#set command, very important.  but it was deprecated. now we work with set_theme
sns.set_theme(style='dark',palette='muted',font='serif', font_scale=1)
sns.barplot(x=['A','B','C'],y=[1,3,2])
plt.show()
```
